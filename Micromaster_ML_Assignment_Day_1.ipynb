```python
#menambahkan kolom "lama_kerja(tahun)" dari table2 ke table1 disebelah kanan kolom "lama_kerja" pada table1
table1['lama_kerja'] = table2['lama_kerja(tahun)']
```

```python
#menampilkan isi dari table1 yang sudah ditambahkan kolom baru dari table2
table1
```

```python
#menambahkan kolom "bonus_selama_kerja" pada table1 disebelah kolom "lama_kerja" yang isinya dikalikan dengan 350000
table1['bonus_selama_kerja'] = table1['lama_kerja'] * 350000
```

```python
#melakukan filtering dengan menampilkan HANYA sales yang lama kerjanya sudah 4 tahun atau lebih
table1[table1['lama_kerja']>=4]
```
